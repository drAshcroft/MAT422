```python
import numpy as np
import time
import matplotlib.pyplot as plt

import random
from scipy.stats import moment

rng = np.random.default_rng()

#create a dataset with a linear relationship

x=np.linspace(0,10,100)

A,B,C=.1,-3,5
Y=C+B*x+A*x**2
plt.plot(x,Y,label='Clean lines')

#add noise to the clean data to show power of linear regression
Y=Y+ rng.standard_normal(len(Y))*5
plt.plot(x,Y,label='Noisy Data')


plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

```

```python
#create a matrix for the input values and linearize the polynomial components
AA = np.vstack([x**x, x, np.ones(len(x))]).T

#perform the linear regression using matrix math
a,b,c=np.dot(np.dot(np.linalg.inv(np.dot(AA.T,AA)), AA.T),Y)

plt.plot(x, Y, 'o', label='Noisy data')
plt.plot(x,a*x**2+ b*x + c, 'r', label='Fitted line')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

print(f'Real parameters   : {A:.2f},{B:.2f},{C:.2f}')
print(f'Derived parameters: {a:.2f},{b:.2f},{c:.2f}')
```

```python
#use numpys built in method to do the math instead of raw code

a,b, c = np.linalg.lstsq(AA, Y, rcond=None) [0]

plt.plot(x, Y, 'o', label='Noisy data')
plt.plot(x,a*x**2+ b*x + c, 'r', label='Fitted line')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

print(f'Real parameters   : {A:.2f},{B:.2f},{C:.2f}')
print(f'Derived parameters: {a:.2f},{b:.2f},{c:.2f}')
```

```python
#use numpy's curve fit method to estimate the parameters

a,b, c = np. polyfit(x,Y,2)

plt.plot(x, Y, 'o', label='Noisy data')
plt.plot(x,a*x**2+ b*x + c, 'r', label='Fitted line')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

print(f'Real parameters   : {A:.2f},{B:.2f},{C:.2f}')
print(f'Derived parameters: {a:.2f},{b:.2f},{c:.2f}')
```
