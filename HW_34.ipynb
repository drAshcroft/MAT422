```python
import matplotlib.pyplot as plt
import glob
import numpy as np
import os
from sklearn.datasets import load_diabetes
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
```

The logistic distribution is a heavy tailed function that can be used in artifical intelligence.  

```python
x=np.linspace(-5,5,1000)
mu=0.0
sigma=1.0

plt.plot(x,1/(sigma*(2*np.pi)**.5)*np.exp(-.5*((x-mu)/sigma)**2),label='normal distribution')
plt.plot(x,.25/sigma*np.cosh((x-mu)/2/sigma)**-2,label='logistic distribution')
plt.xlabel('x')
plt.ylabel('PDF')

plt.legend()
plt.show()
```

The logit function is the quantile function of the logistic distribution.  It is used also the inverse of the logistic function and is used for binary classification throughout machine learning domains.

```python
x=np.linspace(0.001,.9999,1000)
plt.plot(x,np.log(x/(1-x)))
```

We can use the example given by sklearn to perform a simple logistic regression of a toy dataset.  [link text](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

This allows us to test a variety of fitting scenarios on the data for classifying flowers by a number of parameters

```python
X, y = load_iris(return_X_y=True)
clf = LogisticRegression(random_state=0).fit(X, y)
clf.predict(X[:2, :])
clf.predict_proba(X[:2, :])
print('The classification score for this system is: ',clf.score(X, y))
print('Logisitc regression is a good technique for this system')
```

We can reuse this code to see if the logistic regression allows classifcation from more complicated data, like the parameters that are known to predict diabetis

```python

X, y = load_diabetes(return_X_y=True)
clf = LogisticRegression(random_state=0).fit(X, y)
clf.predict(X[:2, :])
clf.predict_proba(X[:2, :])
print('The classification score for this system is: ',clf.score(X, y))
print('Logisitc regression is a bad good technique for this system')
```

The diabetes data is not linear under the logit transform and so cannot be modeled this way to perform classification.  It would likely make sense to use some form of non linear system for the diabetis data.
